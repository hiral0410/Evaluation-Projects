### Handling Null values

```python
# Ckecking skewness before handling missing data
df.skew()
```

Most of the columns have missing values except target and features.

The columns longitude, commissioning year, generation_gwh_2014, generation_gwh_2015, generation_gwh_2016 and generation_gwh_2017, generation_gwh_2018 are numerical type and have huge outliers and skewness,so will replace these columns with median.


```python
# Filling null values using their median values
df['longitude'] = df['longitude'].fillna(df['longitude'].median())
df["commissioning_year"] = df["commissioning_year"].fillna(df["commissioning_year"].median())
df["generation_gwh_2014"] = df["generation_gwh_2014"].fillna(df["generation_gwh_2014"].median())
df["generation_gwh_2015"] = df["generation_gwh_2015"].fillna(df["generation_gwh_2015"].median())
df["generation_gwh_2016"] = df["generation_gwh_2016"].fillna(df["generation_gwh_2016"].median())
df["generation_gwh_2017"] = df["generation_gwh_2017"].fillna(df["generation_gwh_2017"].median())
df["generation_gwh_2018"] = df["generation_gwh_2018"].fillna(df["generation_gwh_2018"].median())
```

The column latitude has numerical data and has no skewness and outliers so we can use mean method.

```python
df["latitude"] = df['latitude'].fillna(df['latitude'].mean())
```

The features other_fuel1 and geolocation_source have categorical data so we can fill the null values using mode method.

```python
# let's check the mode of other_fuel1 
df["other_fuel1"].mode()
```

```python
# let's check the mode of geolocation_source columns
df["geolocation_source"].mode()
```

```python
# Replacing the null valuee by its mode Oil
df["other_fuel1"] = df["other_fuel1"].fillna(df["other_fuel1"].mode()[0])
```

```python
# Filling null values by its mode WRI
df["geolocation_source"] = df["geolocation_source"].fillna(df["geolocation_source"].mode()[0])
```

Now we have removed the null values using imputaion methods and the data is now clean. So we can move further.

### Removing outliers

#### Zscore method

```python
# Features having outliers
features = df[['longitude', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018', 'Power_plant_age']]
```

Now the outliers have been removed.

```python
# Creating new dataframe
new_df = df[(z<3).all(axis = 1)]
new_df
```

This is the new dataframe after removing outliers.

```python
# Checking the dimensions of both the dataframes
print(df.shape)
print(new_df.shape)
```

```python
#Checking data loss
loss = (907-851)/907*100
loss
```

Using zscore method, the data loss is 6.17%. Lets check using IQR.



#### IQR(Inter Qunatile Range) method

```python
# 1st quantile
Q1 = features.quantile(0.25)

# 3rd quantile
Q3 = features.quantile(0.75)

# IQR
IQR = Q3-Q1
df1 = df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis = 1)]
```

Here we have removed outliers using IQR.

```python
# Checking the dimensions of dataframes
print(df.shape)
print(df1.shape)
```

## Modelling

#### Finding the best random state

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
maxAccu = 0
maxRS = 0
for i in range(1, 200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = i)
    mod = RandomForestRegressor()
    mod.fit(x_train, y_train)
    pred = mod.predict(x_test)
    acc = r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print("Maximum r2 score is ", maxAccu, "at Random_state", maxRS)
```

We are getting the maximum r2 score as 86% at random state 185.

### Finding the best random state

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRS = 0
for i in range(1, 200):
    x_df_train, x_df_test, y_df_train, y_df_test = train_test_split(x_df, y_df, test_size = 0.30, random_state = i)
    DTC = DecisionTreeClassifier()
    DTC.fit(x_df_train, y_df_train)
    pred = DTC.predict(x_df_test)
    acc = accuracy_score(y_df_test, pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print("Best Accuracy is", maxAccu, "at random_state", maxRS)
```
